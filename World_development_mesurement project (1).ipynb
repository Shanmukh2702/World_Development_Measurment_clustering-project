```python
# Function to remove outliers based on IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    # Filter the data within IQR bounds
    return df[(df[column] >= (Q1 - 1.5 * IQR)) & (df[column] <= (Q3 + 1.5 * IQR))]

# Applying the function to multiple columns
for column in columns:
    df = remove_outliers_iqr(df, column)
```

```python
# Function to impute outliers with the median
def impute_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Replace outliers with the median
    median = df[column].median()
    df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return df

# Applying the function to multiple columns
for column in columns:
    df = impute_outliers(df, column)
```

```python
# Re-plot boxplots to verify outlier treatment
plt.figure(figsize=(15, 10))
for i, column in enumerate(columns, 1):
    plt.subplot(5, 5, i)
    sns.boxplot(y=df[column])
    plt.title(column)
    plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the Elbow Method   

wcss = []
for i in range(1, 11):
   
     kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
     kmeans.fit(data_new)
     wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# Based on the elbow plot, choose the optimal number of clusters
optimal_num_clusters = 2

# Create the KMeans model
kmeans = KMeans(n_clusters=optimal_num_clusters, init='k-means++', random_state=42)

# Fit the model to the data
kmeans.fit(data_new)
```

```python
# Predict cluster labels for each data point
labels = kmeans.predict(data_new)
```

```python
# Add cluster labels to the original DataFrame
data_new['Cluster'] = labels
```

```python
# Analyze the clusters
print(data_new.groupby('Cluster').mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA  # Or t-SNE

# Reduce dimensionality using PCA (or t-SNE)
pca = PCA(n_components=2)  # Adjust n_components as needed
reduced_data = pca.fit_transform(data_new.drop('Cluster', axis=1))

# Create a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=data_new.iloc[:, 0], y=data_new.iloc[:, 1], hue=data_new['Cluster'], palette='viridis')
plt.title('K-Means Clustering Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Calculate evaluation metrics
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
silhouette_avg = silhouette_score(data_new, labels)
calinski_harabasz_score_value = calinski_harabasz_score(data_new, labels)
davies_bouldin_score_value = davies_bouldin_score(data_new, labels)

print("Silhouette Coefficient:", silhouette_avg)
print("Calinski-Harabasz Index:", calinski_harabasz_score_value)
print("Davies-Bouldin Index:", davies_bouldin_score_value)
```

```python
# Hierarchical Clustering
from sklearn.cluster import KMeans, AgglomerativeClustering,DBSCAN,MeanShift, SpectralClustering, AffinityPropagation
hc = AgglomerativeClustering(n_clusters=2, linkage='ward')  # Adjust linkage criteria
hc.fit(data_new)
labels_hc = hc.labels_
```

```python
print("\nHierarchical Clustering:")
print("Silhouette Score:", silhouette_score(data_new, labels_hc))
print("Calinski-Harabasz Index:", calinski_harabasz_score(data_new, labels_hc))
print("Davies-Bouldin Index:", davies_bouldin_score(data_new, labels_hc))
```
